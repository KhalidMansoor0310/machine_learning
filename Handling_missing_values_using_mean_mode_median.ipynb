```python
# finding percentages of null values at each variable
missing_var_per = df.isnull().sum()/df.shape[0]*100
missing_var_per
```

```python
# finding keys having missing values of percentag greater than 17
df2_drop_cols_gre_17 = missing_var_per[missing_var_per > 17].keys()
df2_drop_cols_gre_17
```
